```python
import numpy as np
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
time = np.linspace(0, 10, 1000)
```

```python
random_signal = np.random.rand(time.shape[0])
```

```python
random_frequencies = np.fft.fft(random_signal)
```

```python
first_part = random_frequencies[:30]
last_part = random_frequencies[-30:]
filtered_frequencies = np.hstack([first_part, np.zeros(time.shape[0] - 60), last_part])
```

```python
filtered_random_signal = np.abs(np.fft.ifft(filtered_frequencies))
```

```python
plt.plot(filtered_random_signal)
```

```python
freq = 20
```

```python
signal = 0.1*np.sin(freq * time) + filtered_random_signal
```

```python
plt.plot(signal)
```

```python
signal_one = signal[:300]
signal_two = signal[300:600]
signal_three = signal[600:900]
```

```python
plt.plot(time[:300], signal_one)
plt.plot(time[300:600], signal_two)
plt.plot(time[600:900], signal_three)
```

```python
def get_merged(signal_one, signal_two):
    signal = np.hstack([signal_one, signal_two])
    frequencies = np.fft.fft(signal)
    frequencies = frequencies[:frequencies.shape[0] // 2]
    x_frequencies = np.arange(len(frequencies))
    
    plt.plot(x_frequencies[0:freq * 3], np.abs(frequencies)[0:freq * 3])
    plt.axvline(freq)
```

```python
get_merged(signal_one, signal_two)
```

```python
get_merged(signal_one, signal_three)
```
